```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 

file1 = pd.read_csv("temposLab107.txt", header = 0, sep='\s+')
file2 = pd.read_csv("temposB710.txt", header = 0, sep='\s+')
file3 = pd.read_csv("temposSequana.txt", header = 0, sep='\s+')

fig1 = plt.figure(figsize=(8,6))

x=np.arange(0,50,2)

axes = fig1.add_axes([0.1,0.1,0.8,0.8])

axes.plot(file1['core'], file1['tempo'], ls='-.', marker='o',markersize=5, label="Lab 107-C")
axes.plot(file2['core'], file2['tempo'], ls='-.', marker='o',markersize=5, label="LNCC - B710")
axes.plot(file3['core'], file3['tempo'], ls='-.', marker='o',markersize=5, label="LNCC - SequanaX")

axes.set_xlabel('Cores')
axes.set_ylabel("Time (s)")

axes.set_title("Comparison of MPI implementation in each case")

plt.grid(linestyle='-', linewidth=0.5)
axes.legend(loc='upper right')

plt.savefig("benchmarkSiestaMPI.png", dpi=400)
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 

file1 = pd.read_csv("temposLab107.txt", header = 0, sep='\s+')
file2 = pd.read_csv("temposB710.txt", header = 0, sep='\s+')
file3 = pd.read_csv("temposSequana.txt", header = 0, sep='\s+')

temposLab107 = file1['tempo']
temposB710 = file2['tempo']
temposSequana = file3['tempo']
eficienciaLab107 = []
eficienciaB710 = []
eficienciaSequana = []

i = 0
while(i < (file1.size)/2):
    eficienciaLab107.append((1 - (temposLab107[i]/temposLab107[0]))*100)
    i = i+1
    
i = 0
while(i < (file2.size)/2):
    eficienciaB710.append((1 - (temposB710[i]/temposB710[0]))*100)
    i = i+1
    
i = 0
while(i < (file3.size)/2):
    eficienciaSequana.append((1 - (temposSequana[i]/temposSequana[0]))*100)
    i = i+1
    

fig1 = plt.figure(figsize=(8,6))

x=np.arange(0,50,2)

axes = fig1.add_axes([0.1,0.1,0.8,0.8])

axes.plot(file1['core'], eficienciaLab107, ls='-.', marker='o',markersize=5, label="Lab 107-C")
axes.plot(file2['core'], eficienciaB710, ls='-.', marker='o',markersize=5, label="LNCC - B710")
axes.plot(file3['core'], eficienciaSequana, ls='-.', marker='o',markersize=5, label="LNCC - SequanaX")

axes.set_xlabel('Cores')
axes.set_ylabel("Efficiency (%)")

axes.set_title("Comparison of MPI efficiency on each case")

plt.grid(linestyle='-', linewidth=0.5)
axes.legend(loc='lower right')

plt.savefig("benchmarkSiestaEficienciaMPI.png", dpi=400)
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 

file1 = pd.read_csv("temposLab107.txt", header = 0, sep='\s+')
file2 = pd.read_csv("temposB710.txt", header = 0, sep='\s+')
file3 = pd.read_csv("temposSequana.txt", header = 0, sep='\s+')

temposLab107 = file1['tempo']
temposB710 = file2['tempo']
temposSequana = file3['tempo']
speedupLab107 = []
speedupB710 = []
speedupSequana = []

i = 0
while(i < (file1.size)/2):
    speedupLab107.append((temposLab107[0]/temposLab107[i]))
    i = i+1
    
i = 0
while(i < (file2.size)/2):
    speedupB710.append((temposB710[0]/temposB710[i]))
    i = i+1
    
i = 0
while(i < (file3.size)/2):
    speedupSequana.append((temposSequana[0]/temposSequana[i]))
    i = i+1
    

fig1 = plt.figure(figsize=(8,6))

x=np.arange(0,50,2)

axes = fig1.add_axes([0.1,0.1,0.8,0.8])

axes.plot(file1['core'], speedupLab107, ls='-.', marker='o',markersize=5, label="Lab 107-C")
axes.plot(file2['core'], speedupB710, ls='-.', marker='o',markersize=5, label="LNCC - B710")
axes.plot(file3['core'], speedupSequana, ls='-.', marker='o',markersize=5, label="LNCC - SequanaX")

axes.set_xlabel('Cores')
axes.set_ylabel("Speedup")

axes.set_title("Comparison of MPI speedup on each case")

plt.grid(linestyle='-', linewidth=0.5)
axes.legend(loc='lower right')

plt.savefig("benchmarkSiestaSpeedupMPI.png", dpi=400)
```
